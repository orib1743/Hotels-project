```python
import numpy as np
import pandas as pd
import sklearn as sk
import sklearn.tree as tree
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime
from collections import Counter
import re
```

```python
data = pd.read_csv('expedia_data.csv')
data.info()
```

```python
# הוספת פיצ'רים נוספים ע"ב משתנים קיימים

# הפונקציה להמרת המחרוזת לתאריך
def convert_to_date(date_str):
    return datetime.strptime(date_str, '%Y-%m-%d')

data['checkin_date'] = data['checkin_date'].apply(convert_to_date)
#data['checkin_day'] = data['checkin_date'].dt.day
data['checkin_month'] = data['checkin_date'].dt.month

# הוספת פיצ'ר ליום בשבוע ע"ב תאריך צקאין
def get_day_of_week(date_obj):
    # קבלת מספר היום בשבוע (0 - ראשון, 1 - שני, וכן הלאה)
    day_of_week = date_obj.weekday()
    return day_of_week

# החלפת הערכים בעמודה 'bed_type' לערכי הסיווג החדש
data['checkin_day_week'] = data['checkin_date'].apply(get_day_of_week)
```

```python
def add_all_services_column(data):
    # הגדרת תנאי שמבצע בדיקה אם כל השירותים מוצעים על ידי המלון
    conditions = (data['reserve_now_pay_later'] == 1) & \
                 (data['fully_refundable'] == 1) & \
                 (data['free_cancellation'] == 1) & \
                 (data['breakfast'] == 1)
    
    # הוספת העמודה החדשה עם תוצאה של 1 אם כל התנאים מתקיימים, אחרת 0
    data['all_services'] = conditions.astype(int)
    return data

# יישום הפונקציה על הנתונים
data = add_all_services_column(data)

# הצגת מספר שורות ראשונות של הדאטה לאחר הוספת העמודה החדשה
data.head()
value_counts = data['all_services'].value_counts()
value_counts
```

```python
price_correlations = data[['price','all_services',  'checkin_day_week']].corr()['price'].sort_values(ascending=False)

print("Correlations with price:")
print(price_correlations)

plt.figure(figsize=(10, 6))
sns.barplot(x=price_correlations.index, y=price_correlations.values)
plt.title('Correlation of Features with Price')
plt.xlabel('Features')
plt.ylabel('Correlation with Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Algorithm -LinearRegression

#### Split the data to training set and test set.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


data_types = [data_min_max_scaled, data_wo_features]

for df in data_types:
    if df is data_wo_features:
        print("Working on: data_wo_features")
    else:
        print("Working on: data")
    
    X = df.drop('price', axis=1)
    y = df['price']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
    
    #scaler = MinMaxScaler()
    #X_train = scaler.fit_transform(X_train)
    #X_test = scaler.transform(X_test)
    
    lr = LinearRegression()
    lr.fit(X_train,y_train)

    print(f'Intercept: {lr.intercept_}')
    
    pd.options.display.float_format = '{:.3f}'.format
    # נכניס את הנתונים לטבלת DF
    cdf = pd.DataFrame(lr.coef_, X_train.columns, columns=['Coeff'])
    print(cdf)
        
    #### Predict the test se-t
    pred_test = lr.predict(X_test)
    pred_train = lr.predict(X_train)

    #### Print a scatterplot of the real y values of the test set against the predicted y values
    plt.scatter(pred_train, y_train-pred_train, c='b',alpha=0.15)
    plt.scatter(pred_test, y_test-pred_test, c='r',alpha=0.15)
    plt.hlines(y=0,xmin=-50,xmax=500)
    plt.show()

    plt.scatter(y_test, pred_test)
    plt.xlabel('Y test')
    plt.ylabel('Predicted Y')
    plt.show()

    #### Print the relevant evaluation metrics
    print('Test MAE: ', "%.4f" % mean_absolute_error(y_test, pred_test))
    print('Test MSE: ', "%.4f" % mean_squared_error(y_test, pred_test))
    print('Test RMSE: ', "%.4f" % np.sqrt(mean_squared_error(y_test, pred_test)))
    print('Test R2: ', "%.4f" % r2_score(y_test, pred_test))
    print()
    print('Train MAE: ', "%.4f" % mean_absolute_error(y_train,pred_train))
    print('Train MSE: ', "%.4f" % mean_squared_error(y_train,pred_train))
    print('Train RMSE: ', "%.4f" % np.sqrt(mean_squared_error(y_train,pred_train)))
    print('Train R2: ', "%.4f" % r2_score(y_train,pred_train))
    print()

    #### Print a histogram the residuals
    sns.distplot(y_test-pred_test)
```

## Conclusion
ערכי המקדמים ("קופישיאנט") מציינים את הגודל ואת הכיוון של ההשפעה של כל משתנה על המשתנה התלותי (מחיר)

LOS - משפיע בצורה חיובית על המחיר

# Algorithm - Decision Tree Regressor

```python
from sklearn import tree
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix, classification_report, roc_curve
from IPython.display import Image
import pydotplus
```

```python
data_types = [data,data_min_max_scaled, data_wo_features]
depth_values = [2, 3, 5, 8, 10]

for df in data_types:
    if df is data_wo_features:
        print("Working on: data_wo_features")
    elif df is data_min_max_scaled:
        print("Working on: data_min_max_scaled")
    else:
        print("Working on: data")
    
    X = df.drop('price', axis=1)
    y = df['price']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
    for depth in depth_values:
        print(f"Max_depth: {depth}")
        print()
        lr2 = tree.DecisionTreeRegressor(max_depth=depth)
        lr2.fit(X_train, y_train)
        
        #### Predict the test set
        pred_train = lr2.predict(X_train)
        pred_test = lr2.predict(X_test)
        
        #### Print a scatterplot of the real y values of the test set against the predicted y values
        plt.scatter(pred_train, y_train-pred_train, c='b',alpha=0.15)
        plt.scatter(pred_test, y_test-pred_test, c='r',alpha=0.15)
        plt.hlines(y=0,xmin=0,xmax=5000)
        plt.show()
        
        print('Test MAE: ', "%.4f" % mean_absolute_error(y_test, pred_test))
        print('Test MSE: ', "%.4f" % mean_squared_error(y_test, pred_test))
        print('Test RMSE: ', "%.4f" % np.sqrt(mean_squared_error(y_test, pred_test)))
        print('Test R2: ', "%.4f" % r2_score(y_test, pred_test))
        print()
        print('Train MAE: ', "%.4f" % mean_absolute_error(y_train,pred_train))
        print('Train MSE: ', "%.4f" % mean_squared_error(y_train,pred_train))
        print('Train RMSE: ', "%.4f" % np.sqrt(mean_squared_error(y_train,pred_train)))
        print('Train R2: ', "%.4f" % r2_score(y_train,pred_train))
        print()
        
        #### Print a histogram the residuals
        plt.figure(figsize=(10, 6))  # יצירת חלון גרפים חדש לכל גרף
        sns.distplot(y_test-pred_test, kde=False)
        plt.title(f"Residuals Distribution (Max_depth: {depth})")
        plt.xlabel('Residuals')
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()

```

## Conclusion
ככל שעומק העץ גדול יותר ניתן לראות שהתוצאות משתפרות והמודל טוב יותר אולם ייתכן שמדובר בהתאמת יתר ולכן, נבדוק עומקים נוספים וננסה להגביל את עומק העץ

כאשר עומק העץ 2, 3 או 5, התקבלו תוצאות פחות טובות מאלה של הרגרגסיה הלינארית

הרצנו מס' פעמים, ננסה גם לבצע חלוקה שונה ולהגדיל את הטסט על חשבון הטריין ולבחון את ההשפעה

## Split the data to training set and test set

```python
data_types = [data, data_min_max_scaled, data_wo_features]
test_size_values = [0.2, 0.4, 0.5]

for df in data_types:
    if df is data_wo_features:
        print("Working on: data_wo_features")
    elif df is data_min_max_scaled:
        print("Working on: data_min_max_scaled")
    else:
        print("Working on: data")
    
    X = df.drop('price', axis=1)
    y = df['price']
    
    for size in test_size_values:
        print(f"Test_size: {size}")
        print()
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=101)
        lr2 = tree.DecisionTreeRegressor(max_depth=5)
        lr2.fit(X_train, y_train)
        
        #### Predict the test set
        pred_train = lr2.predict(X_train)
        pred_test = lr2.predict(X_test)
        
        #### Print a scatterplot of the real y values of the test set against the predicted y values
        plt.scatter(pred_train, y_train-pred_train, c='b',alpha=0.15)
        plt.scatter(pred_test, y_test-pred_test, c='r',alpha=0.15)
        plt.hlines(y=0,xmin=0,xmax=5000)
        plt.show()
        
        print('Test MAE: ', "%.4f" % mean_absolute_error(y_test, pred_test))
        print('Test MSE: ', "%.4f" % mean_squared_error(y_test, pred_test))
        print('Test RMSE: ', "%.4f" % np.sqrt(mean_squared_error(y_test, pred_test)))
        print('Test R2: ', "%.4f" % r2_score(y_test, pred_test))
        print()
        print('Train MAE: ', "%.4f" % mean_absolute_error(y_train,pred_train))
        print('Train MSE: ', "%.4f" % mean_squared_error(y_train,pred_train))
        print('Train RMSE: ', "%.4f" % np.sqrt(mean_squared_error(y_train,pred_train)))
        print('Train R2: ', "%.4f" % r2_score(y_train,pred_train))
        print()
        
        #### Print a histogram the residuals
        plt.figure(figsize=(10, 6))  # יצירת חלון גרפים חדש לכל גרף
        sns.distplot(y_test-pred_test, kde=False)
        plt.title(f"Residuals Distribution (Test Size: {size})")
        plt.xlabel('Residuals')
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()

```
